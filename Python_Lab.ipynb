**Factorial**

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = factorial_recursive(num)
    print("The factorial of", num, "is", result)
```

**Write a program to display all prime numbers within a range**

```python
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(f"Prime numbers in the range {start} to {end}:")
for num in range(start, end + 1):
    if is_prime(num):
        print(num)
```

**13. Sum of all odd and even number within a given range**

```python
def sum_odd_even_numbers(start, end):
    sum_even = 0
    sum_odd = 0
    odd_numbers = []
    even_numbers = []

    for number in range(start, end + 1):
        if number % 2 == 0:
            sum_even += number
            even_numbers.append(number)
        else:
            sum_odd += number
            odd_numbers.append(number)

    return sum_odd, sum_even, odd_numbers, even_numbers

start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))

odd_sum, even_sum, odd_numbers, even_numbers = sum_odd_even_numbers(start_range, end_range)
print("Odd numbers in the range:", odd_numbers)
print(f"Sum of odd numbers: {odd_sum}")
print("Even numbers in the range:", even_numbers)
print(f"Sum of even numbers: {even_sum}")
```

**15. To find if a number is amrmstrong number or not**

```python
def is_armstrong_number(number):
    num_str = str(number)
    n = len(num_str)
    sum_of_powers = 0
    for digit_str in num_str:
        digit = int(digit_str)
        sum_of_powers += digit ** n
    if sum_of_powers == number:
        return True
    else:
        return False

num = int(input("Enter a number: "))
if is_armstrong_number(num):
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")
```

**Write a program to find the largest & smallest num in a list**

```python
def find_largest_and_smallest(numbers):
    if not numbers:
        return None, None  # Handle empty list case

    largest = smallest = numbers[0]  # Initialize with the first element

    for num in numbers:
        if num > largest:
            largest = num
        elif num < smallest:
            smallest = num

    return largest, smallest

# Example usage:
numbers = [12, 45, 6, 72, 31, 50, 8]
largest, smallest = find_largest_and_smallest(numbers)

if largest is not None and smallest is not None:
    print(f"Largest number: {largest}")
    print(f"Smallest number: {smallest}")
else:
    print("The list is empty.")
```

```python
def is_perfect_number(number):
    sum_of_divisors = 1

    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            sum_of_divisors += i
            if i != number // i:
                sum_of_divisors += number // i

    return sum_of_divisors == number

number = int(input("Enter a number to check if it's a perfect number: "))

if is_perfect_number(number):
    print(f"{number} is a perfect number.")
else:
    print(f"{number} is not a perfect number.")
```

```python
class Person:
    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("City:", self.city)

# Example usage:
person1 = Person("Yash", 22, "Mumbai")
person1.display_details()
```
